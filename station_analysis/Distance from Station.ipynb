```python
from datetime import datetime
from meteostat import Stations, Hourly
import pickle
from geopy import distance
import pandas
import os
```

```python
ROOT_DIR = os.path.realpath(os.path.join(os.getcwd(), '..'))
st_loc = os.path.join(ROOT_DIR, 'station_analysis')
st_loc
```

```python
#these are all the stations that are in the lat/long bounds for the weathersmallraw file.  
stations = Stations()
stations = stations.bounds((50, -100), (40, -80))
stationdf=stations.fetch()
#print('Stations in northern hemisphere:', stations.count())
```

```python
#Function that can take in a station code, lat and long of another station and return the distance in miles.
def MilesBetweenStations(station,latitude, longitude):
    return distance.distance((stationdf.loc[station]["latitude"],stationdf.loc[station]["longitude"]),(latitude,longitude)).miles
```

```python
station = 'KARB0'
#calculating the distance for Ann Arbor.  
stationdf["milesfromstat"]=stationdf.apply(lambda x: MilesBetweenStations(station,x.latitude, x.longitude),axis=1)
```
