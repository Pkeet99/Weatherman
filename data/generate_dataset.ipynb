```python
import pandas as pd
from meteostat import Stations
import os
```

```python
stations = Stations()
stations = stations.bounds((50, -110), (30, -65))
stationlist=stations.fetch()
```

```python
stationlist.groupby(['country','region']).count()
```

```python
if not os.path.isdir(r'data/cache'):
    os.makedirs(r'data/cache')
```

```python
from datetime import datetime
from meteostat import Hourly

Hourly.cache_dir = r'data/cache'

# Set time period
start = datetime(2019, 1, 1)
end = datetime(2022, 8, 31, 23, 59)

# Get hourly data
data = Hourly(stationlist, start, end)
data = data.fetch()
data
```

```python
data.to_pickle('data/weatherraw.pkl')


```
