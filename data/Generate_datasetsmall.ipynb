```python
from datetime import datetime
from meteostat import Stations, Hourly
import pickle
```

```python
stations = Stations()
stations = stations.bounds((50, -100), (35, -80))
stationlist=stations.fetch()
#print('Stations in northern hemisphere:', stations.count())

stationlist.drop(['NO755','NNY2V','O879F'],errors='ignore',inplace=True)
stationlist
```

```python
start = datetime(2021, 10, 15)
end = datetime(2022, 10, 15, 23, 59)
```

```python
#print(stationlist.to_string())
```

```python
Hourly.cache_dir = r'data/cache'
data = Hourly(stationlist, start=start, end=end)
```

```python
#data = Hourly('72219', start, end)
data = data.fetch()
```
