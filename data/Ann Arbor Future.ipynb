```python
import pandas as pd
from meteostat import Stations
from datetime import datetime
import os
import pickle
```

```python
from meteostat import Hourly

#Hourly.chunked = False


start = datetime(2022, 9, 1)
end = datetime(2022, 10, 15, 23, 59)
data = Hourly('KARB0',start=start, end=end)
data = data.fetch()
```

```python
data
```

```python
data.drop(columns=['snow', 'wpgt','tsun','coco'],inplace=True)
```

```python
ROOT_DIR = os.path.realpath(os.path.join(os.getcwd(), '..'))

data.to_pickle(r'data/weatherAAtest.pkl')
```

```python
data.reset_index(inplace=True)
data['timediff']=data['time'].diff(periods=1) #difference for hours
data['timediff']=data['timediff']/pd.Timedelta(hours=1) #converted to hour int
```
