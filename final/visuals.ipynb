```python
import altair as alt
import pandas as pd
import os
from sktime.performance_metrics.forecasting import mean_absolute_error
```

```python
ROOT_DIR = os.path.realpath(os.path.join(os.getcwd(), '..'))
xt = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations.csv')
lasso = os.path.join(ROOT_DIR, 'final','Lasso_results_other_stations.csv')
tw = os.path.join(ROOT_DIR, 'final','Tweedie_results_other_stations.csv')
sarima = os.path.join(ROOT_DIR, 'final','SARIMA_Results.csv')
sarima_wf = os.path.join(ROOT_DIR, 'final','SARIMA_Results_wf.csv')

xt_km = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial_cluster).csv')
xt_km_wf = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial clusters-walk forward).csv')
xt_km2 = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial_cluster2).csv')
xt_km2_wf = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial_cluster2-wf).csv')

xt_db = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(db_scan_cluster).csv')
xt_db_wf = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(db_scan_cluster-wf).csv')

xt_db_km = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(db_km_scan_cluster).csv')
dummy = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(dummy).csv')
dummy_wf = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(dummy-wf).csv')

results = os.path.join(ROOT_DIR, 'Results','Extra_Tree_results.csv')
res = pd.read_csv(results,parse_dates=['time'])
res.columns = ('time', 'xt(aa_only)','actual')
res = res[['time','actual', 'xt(aa_only)']]

files = [(xt,'xt'), (lasso,'lasso'), (tw,'tw'), (sarima,'sarima'), (sarima_wf,'sarima_wf'),
         (xt_km,'xt_km'), (xt_km_wf,'xt_km_wf'), (xt_km2,'xt_km2'), (xt_km2_wf,'xt_km2_wf'),
         (xt_db,'xt_db'), (xt_db_wf,'xt_db_wf'), (xt_db_km,'xt_db_km'), 
         (dummy,'dummy'), (dummy_wf,'dummy_wf')]

for file,name in files:
    df = pd.read_csv(file,parse_dates=[0])
    try:
        df.columns = ('time', name)
    except:
        df.columns = ('time', name,'act')
        df = df.drop(columns='act')
    res = res.merge(df, on='time',how='outer')
res
```

```python
dum_dum = os.path.join(ROOT_DIR, 'unsupervised_models','kmeansclustering2.csv')
dum_dum = pd.read_csv(dum_dum)

name = dum_dum['name'][1]
dum_dum = dum_dum['station'][1]

dummy_station = os.path.join(ROOT_DIR, 'data_cleaning','cleanweather.pkl')
dummy_station = pd.read_pickle(dummy_station)
dummy_station = dummy_station[(dummy_station['station']==dum_dum)&(dummy_station['time']>'2022-08-31 23')][['time','temp']]
dummy_station.set_index('time',inplace=True)
dummy_df = pd.DataFrame()
dummy_df['marshall'] = dummy_station['temp'].astype(float)
dummy_df = dummy_df.merge(res,on='time')[['time','marshall','actual']]
dummy_df
```

```python
dummy_mae = round(mean_absolute_error(dummy_df['marshall'], dummy_df['actual']), 3)
dummy_mae
```

```python
dummy_df.melt(id_vars='time')
```

```python
alt.data_transformers.disable_max_rows()

domain = ['actual','marshall']
range_ = ['cornflowerblue', 'darkred']

line = alt.Chart(dummy_df.melt(id_vars='time')).mark_line(size=1).encode(
    x=alt.X('time:T', axis=alt.Axis(tickSize=0, labelFontSize=12),title=None),
    y=alt.Y('value:Q',title='Temperature (C°)'),
    color=alt.Color('variable:N', 
                    legend=alt.Legend(title='Legend'),
                    scale=alt.Scale(domain=domain, range=range_)),
    strokeDash=alt.StrokeDash('variable',legend=None)
).properties(
    height=400,
    width=800,
    title={
      "text": ['Marshall, MN  Model Comparisons'], 
      "subtitle": ["24hr offset from Ann Arbor Sept 1 - Oct 15"],
      "color": "black",
      "subtitleColor": "gray"
    }
)
text = alt.Chart({'values':[{'x': '2022-10-11 19:00:00', 'y': 30}]}).mark_text(
    text=f"MAE: 5.328",
    color='red',
    size=15
).encode(x='x:T', y='y:Q')
line+text
```

```python

models = [col for col in res.columns if col not in ('time','actual')]
mae_dict = {}
for col in models:
    cur_mae = round(mean_absolute_error(res[col], res['actual']), 3)
    print(f"{col} MAE = {cur_mae}")
    mae_dict[col] = cur_mae
```
