```python
import altair as alt
import pandas as pd
import os
```

```python
ROOT_DIR = os.path.realpath(os.path.join(os.getcwd(), '..'))
xt_db_km = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(db_km_scan_cluster).csv')
xt_km = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial_cluster).csv')
xt = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations.csv')
xt_db = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(db_scan_cluster).csv')
xt_km_wf = os.path.join(ROOT_DIR, 'final','Extra_Tree_results_other_stations(initial clusters-walk forward).csv')
results = os.path.join(ROOT_DIR, 'Results','Extra_Tree_results.csv')
res = pd.read_csv(results)
res.columns = ('time', 'aa_only','actual')
res

files = [(xt,'xt'), (xt_km,'xt_km'), (xt_km_wf,'xt_km_wf'), (xt_db,'xt_db'), (xt_db_km,'xt_db_km')]
for file,name in files:
    df = pd.read_csv(file)
    df.columns = ('time', name)
    res = res.merge(df, on='time',how='outer')

res
```

```python
res.dropna(inplace=True)
res.melt(id_vars='time')
```

```python
alt.data_transformers.disable_max_rows()
alt.Chart(res[['time','actual','aa_only']].melt(id_vars='time')).mark_line(size=1).encode(
    x=alt.X('time:T', axis=alt.Axis(tickSize=0, labelFontSize=12),title=None),
    y=alt.Y('value:Q',title='Temperature'),
    color=alt.Color('variable',legend=alt.Legend(title='Model')),
    strokeDash='variable'
).properties(
    height=400,
    width=800,
    title={
      "text": ['Extra Trees Regression Model Comparisons'], 
      "subtitle": ["Ann Arbor only Sept 1 - Oct 15"],
      "color": "black",
      "subtitleColor": "gray"
    }
)
```
